```python
import matplotlib.pyplot as plt

epochs = [1, 2, 3, 4, 5]

train = [60, 68, 75, 82, 90]
validation = [58, 66, 73, 79, 85]

plt.plot(epochs, train, label="Training Accuracy")

plt.plot(epochs, validation, label="Validation Accuracy")

plt.title("Model Accuracy")

plt.xlabel("Epoch")

plt.ylabel("Accuracy (%)")

plt.legend(
    loc="best",
    fontsize=10,
    title="Accuracy Curves")
print(plt.style.available)
plt.style.use('dark_background')
plt.grid(
    color="gray",
    linestyle="--",
    linewidth=0.8,
    alpha=0.5
)


plt.show()
```

```python
#understanding types of ploting graphs using f1 racing as an exaample

import matplotlib.pyplot as plt

drivers = ["Verstappen", "Norris", "Leclerc", "Russell", "Hamilton"]

points = [25, 18, 15, 12, 10]

avg_speed = [228, 225, 223, 221, 220]

pit_stops = [2, 2, 3, 2, 3]

laps = [1,2,3,4,5,6,7,8]

lap_time = [92.1,91.8,91.4,91.0,90.8,90.9,91.1,91.3]

plt.figure(figsize=(8,5))

plt.plot(
    laps,
    lap_time,
    marker="o",
    linewidth=4,
    color='red'
)

plt.title('laps vs lap_time')
plt.xlabel('laps')
plt.ylabel('lap_times')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
#scatter plot

plt.figure(figsize=(10,10))
plt.scatter( drivers,
            points,
             s=120,
             color='blue'
)
plt.title('drivers vs points')
plt.xlabel('drivers')
plt.ylabel('points')
plt.grid(True)
plt.style.use('dark_background')
plt.legend(loc='best')
plt.show()
```

```python
#bar chart
plt.figure(figsize=(8,8))
plt.barh(
    drivers,
    points,
    color='red'
)
plt.title('drivers vs points')
plt.xlabel('drivers')
plt.ylabel('points')
plt.grid(True)
plt.style.use('dark_background')
plt.legend(loc='best'
)
plt.show()
```

```python
#histogram
plt.figure(figsize=(9,8))
plt.hist(
    lap_time,
    bins=5,
    color='green'

)
plt.title('lap_time distribution')
plt.xlabel('lap_time')
plt.ylabel('Frequency')
plt.grid(True)
plt.style.use('dark_background')
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt

compound = ["Soft", "Medium", "Hard"]
laps = [20, 40, 10]

plt.figure(figsize=(6,6))

plt.pie(
    laps,
    labels=compound,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("F1 Tyre Compound Usage")

plt.show()
```

```python
import matplotlib.pyplot as plt

compound = ["Soft","Medium","Hard"]

laps = [20,40,10]

colors = ["#E10600","#FFD700","#808080"]

explode = [0.08,0,0]

plt.figure(figsize=(7,7))

plt.pie(
    laps,
    labels=compound,
    colors=colors,
    explode=explode,
    autopct="%1.1f%%",
    startangle=90,
    shadow=True,
    wedgeprops={
        "edgecolor":"white",
        "linewidth":2
    },
    textprops={
        "fontsize":12,
        "fontweight":"bold"
    }
)

plt.title(
    "F1 Tyre Compound Usage",
    fontsize=16,
    fontweight="bold"
)

plt.show()
```

```python
import matplotlib.pyplot as plt

lap_time=[92.1,91.8,91.4,91.0,90.8,90.9,91.1,91.3]

drivers=["Max","Lando","Charles","George"]

points=[25,18,15,12]

speed=[228,225,223,221]

plt.figure(figsize=(10,8))

plt.subplot(2,2,1)

plt.plot(lap_time)

plt.title("Lap Time")


plt.subplot(2,2,2)

plt.bar(drivers,points)

plt.title("Points")
plt.subplot(2,2,3)

plt.scatter(speed,points)

plt.title("Speed vs Points")

plt.subplot(2,2,4)

plt.hist(lap_time)

plt.title("Lap Distribution")

plt.tight_layout()

plt.show()
```

```python
#subplots with ax and fig

import matplotlib.pyplot as plt

drivers=["Max","Lando","Charles","George"]

points=[25,18,15,12]

speed=[228,225,223,221]
lap=[92.1,91.8,91.4,91.0]
fig,ax=plt.subplots(
    2,
    2,
    figsize=(10,8)
)

ax[0,0].plot(lap)
ax[0,0].set_title("Lap Time")

ax[0,1].bar(drivers,points)

ax[0,1].set_title("Points")
ax[1,0].scatter(speed,points)
ax[1,0].set_title("Speed vs Points")

ax[1,1].hist(lap)
ax[1,1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

```python
#object oriented api using axis and figure
fig, ax = plt.subplots(2,2, figsize=(10,8))

ax[0,0].plot(lap)

ax[0,0].set_title("Lap")
ax[0,1].bar(drivers, points)
ax[0,1].set_title("Points")
ax[1,0].scatter(speed, points)

ax[1,0].set_title("Scatter")
ax[1,1].hist(lap)

ax[1,1].set_title("Histogram")
plt.tight_layout()
plt.style.use("Solarize_Light2")
plt.show()
```
