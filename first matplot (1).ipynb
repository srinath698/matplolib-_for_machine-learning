```python
#title and labels
import matplotlib.pyplot as plt
a=[1,2,3]
b=[3,4,5]
plt.plot(a,b)
plt.title('random numbers')
plt.xlabel('x-random')
plt.ylabel('y-random')
plt.xticks([1,2,3,4])
plt.yticks([1,2,3,4])
plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.show()
```

```python
month=[1,2,3,4]
sales=[100,103,200,201]
plt.plot(month,sales)
plt.show()

plt.show()
plt.xticks([1,2,3,4],['jan','feb','sep','aug'])
plt.xticks(fontsize=11)
plt.yticks(fontsize=10)
plt.xticks(color='red')
plt.yticks(color='blue')
plt.show()
```

```python
#customizing titles and labels
months = ["Jan", "Feb", "Mar", "Apr"]
sales = [120, 180, 200, 250]
plt.plot(months, sales)
plt.title("every month sales"
,fontsize=18,color='red',fontfamily='cursive',fontweight='bold',pad=20)
plt.xlabel('months',fontsize=12,color='blue',fontfamily='cursive',fontweight='bold',labelpad=20)
plt.ylabel('sales',fontsize=12,color='green',fontfamily='cursive',fontweight='bold',labelpad=20)
plt.show()
```

```python
x=[1,2,3,4]
dsa_marks=[90,92,96,98]
ml_marks=[93,95,98,100]
plt.plot(x,dsa_marks,label='dsa_marks',color='red')
plt.plot(x,ml_marks,label='ml_marks',color='green')
plt.title('marks of ml and dsa',color='yellow',fontsize=12)
plt.xlabel('no of hrs studied')
plt.ylabel('marks')
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt

epochs = [1, 2, 3, 4, 5]

train = [60, 68, 75, 82, 90]
validation = [58, 66, 73, 79, 85]

plt.plot(epochs, train, label="Training Accuracy")

plt.plot(epochs, validation, label="Validation Accuracy")

plt.title("Model Accuracy")

plt.xlabel("Epoch")

plt.ylabel("Accuracy (%)")

plt.legend(
    loc="best",
    fontsize=10,
    title="Accuracy Curves")
print(plt.style.available)
plt.style.use('dark_background')
plt.grid(
    color="gray",
    linestyle="--",
    linewidth=0.8,
    alpha=0.5
)


plt.show()
```
